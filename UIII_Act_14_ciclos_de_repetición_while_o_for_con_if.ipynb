```python
# 26/mayo/2026

# Triángulo Rectángulo:
# Programa que dibuja un triángulo rectángulo
# de asteriscos de altura N (creciente).

# ======================================================
# 🔺 CBTis 128
# 📚 Programación Lógica 2026
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Ing. Eliseo Nava
# 👨‍🎓 Alumno: Axel Guzman NL = 27
# 🔺 Proyecto: Triángulo de Asteriscos
# ======================================================

print("======================================")
print("   TRIANGULO RECTANGULO DE ASTERISCOS")
print("======================================")

altura = int(input("Ingresa la altura del triángulo: "))

# --------------------------------------
# Validación con while
# --------------------------------------

while altura <= 0:

    print("⚠️ La altura debe ser mayor a 0")
    altura = int(input("Ingresa nuevamente la altura: "))

print("\n=========== TRIANGULO ===========\n")

# --------------------------------------
# Ciclo while
# --------------------------------------

fila = 1

while fila <= altura:

    columna = 1

    while columna <= fila:

        print("*", end=" ")

        columna = columna + 1

    print()

    fila = fila + 1

# --------------------------------------
# Resultado final
# --------------------------------------

print("\n======================================")
print("✅ Triángulo generado correctamente")
print("👨‍💻 Programa realizado por Axel Guzman")
print("======================================")
```
